```python
<p><
<h3>Cómo usar el proyecto?</h3>
<br>
<h6>Una vez ejecutado el jupyter notebook, el usuario debe ingresar el nombre de una comuna de la región metropolitana 
de Santiago. Luego de unos segundos, el programa entregará la siguiente información sobre los combustibles de dicha comuna:</h6>
<br>
<ul>
<li>Comuna: </li>
<li>Fecha y hora de actualización: </li>
<li>Horario de atención:</li>
<li>Nombre distribuidor:</li>
<li>Razón Social: </li>
<li>Dirección:</li>
<li>gasolina 93:</li>
<li>gasolina 95:</li>
<li>gasolina 97:</li>
<li>petroleo diesel:</li>
<li>efectivo:</li>
<li>cheque:</li>
<li>tarjetas bancarias:</li>
<li>tarjetas grandes tiendas:</li>
<li>tienda:</li>
<li>farmacia: </li>
<li>mantencion:</li>
<li>autoservicio:</li>
</ul>
<h6>El enfoque de este proyecto está en presentar una comparación entre los precios de los distintos tipos de combustible vehicular 
de una comuna seleccionada por el usuario. Es por esto que se señalan conclusiones respecto a los valores de los combustibles.</h6>
<br>
<h6>Módulos ocupados en el desarrollo de este proyecto:</h6>
<ul>
<li>requests</li>
</ul>
<h6>Para finalizar, el proyecto entrega un registro de toda la información de una comuna en un archivo txt ("archivo.txt"). En caso que
se pierda la conexión de internet, se podrá contar con este registro. Dicho archivo se sobreescribe cada vez que se ejecuta el proyecto.
</h6>
```

```python
import requests

token = "hISHc9TmCE"
url = "http://api.cne.cl/v3/combustibles/vehicular/estaciones"
comuna = input('Ingrese la comuna: ').title()

res = requests.get(f"{url}?token={token}")
res_json = res.json()

archivo = open("archivo.txt", "w")

lista_precio = []
barato = []
caro = []
gas_93 = []
gas_95 = []
gas_97 = []
diesel = []
glp = []
gnc = []

for d in res_json['data']:
    if d['id_region']=='13' and d['nombre_comuna']==comuna:
        print('Comuna:', d['nombre_comuna'])
        archivo.write("\n")
        archivo.write(f"Comuna: {d['nombre_comuna']}"+ "\n")
        print('\tFecha actualización:', d['fecha_hora_actualizacion'])
        archivo.write(f"\tFecha y hora de actualización: {d['fecha_hora_actualizacion']}"+ "\n")
        print('\tHorario de atención:', d['horario_atencion'])
        archivo.write(f"\tHorario de atención: {d['horario_atencion']}"+ "\n")
        print('\tNombre distribuidor:', d['distribuidor']['nombre'])
        archivo.write(f"\tNombre distribuidor: {d['distribuidor']['nombre']}"+ "\n")
        print('\tRazón Social:', d['razon_social'])
        archivo.write(f"\tRazón Social: {d['razon_social']}"+ "\n")
        print('\tDirección:', d['direccion_calle'], d['direccion_numero'])
        archivo.write(f"\tDirección: {d['direccion_calle']} {d['direccion_numero']}"+ "\n")
        for nombre, valor in (d['precios']).items():
            if nombre == 'gasolina 93':
                gas_93.append((nombre, valor))
            elif nombre == 'gasolina 95':
                gas_95.append((nombre, valor))
            elif nombre == 'gasolina 97':
                gas_97.append((nombre, valor))
            elif nombre == 'petroleo diesel':
                diesel.append((nombre, valor))
            elif nombre == 'glp vehicular':
                glp.append((nombre, valor))
            elif nombre == 'gnc':
                gnc.append((nombre, valor))
            print(nombre, ':', valor)
            archivo.write(f"{nombre}: {valor}"+ "\n")
            lista_precio.append(valor)
            lista_precio.sort()
            if valor==lista_precio[0]:
                barato.append((valor, nombre))
                barato.sort()
            elif valor ==lista_precio[-1]:
                caro.append((valor,nombre))
                caro.sort()
        for met, res in (d['metodos_de_pago']).items():
            if res == False:
                res = 'No'
            else:
                res = 'Sí'
            print('\tMétodos de pago:', met, res)
            archivo.write(f"\t{met}: {res}"+ "\n")
        for s, servicio in (d['servicios']).items():
            if servicio == False:
                servicio = 'No'
            else:
                servicio = 'Sí'
            print('\tServicios:', s, servicio)
            archivo.write(f"\t{s}: {servicio}"+ "\n")
           
        print()

gas_93.sort()
gas_95.sort()
gas_97.sort()
diesel.sort()
glp.sort()
gnc.sort()

try:
# conclusiones
    archivo.write("\n")
    print("=====================================================================================================")
    archivo.write("====================================================================================================="+ "\n")
    print('Basado en la información consultada, se puede concluir que: \n')
    archivo.write("Basado en la información consultada, se puede concluir que:" +"\n")
    print(f"\tEl combustible más barto de la comuna de {comuna} es {(barato[0][1])} y vale: {lista_precio[0]} pesos por litro.")
    archivo.write(f"El combustible más barto de la comuna de {comuna} es {(barato[0][1])} y vale: {lista_precio[0]} pesos por litro." +"\n")
    print(f"\tEl combustible más caro de la comuna de {comuna} es {caro[-1][-1]} y vale: {lista_precio[-1]} pesos por litro.")
    archivo.write(f"El combustible más caro de la comuna de {comuna} es {caro[-1][-1]} y vale: {lista_precio[-1]} pesos por litro." +"\n")
    print(f"\tLa diferencia entre el combustible más barto y más caro es: {lista_precio[-1] - lista_precio[0]} pesos.")
    archivo.write(f"La diferencia entre el combustible más barto y más caro es: {lista_precio[-1] - lista_precio[0]} pesos." +"\n")
    print()
    archivo.write("\n")
    print("Comparación de precios entre tipos de combustibles: \n")
    archivo.write("Comparación de precios entre tipos de combustibles: "+"\n")
    print(f"\tGasolina 93 más barato: {gas_93[0][1]}. Más caro: {gas_93[-1][-1]}. Hay una diferencia de: {(gas_93[-1][-1])-(gas_93[0][1])}")
    archivo.write(f"\tGasolina 93 más barato: {gas_93[0][1]}. Más caro: {gas_93[-1][-1]}. Hay una diferencia de: {(gas_93[-1][-1])-(gas_93[0][1])}"+"\n")
    print(f"\tGasolina 95 más barato: {gas_95[0][1]}. Más caro: {gas_95[-1][-1]}. Hay una diferencia de: {(gas_95[-1][-1])-(gas_95[0][1])}")
    archivo.write(f"\tGasolina 95 más barato: {gas_95[0][1]}. Más caro: {gas_95[-1][-1]}. Hay una diferencia de: {(gas_95[-1][-1])-(gas_95[0][1])}"+"\n")
    print(f"\tGasolina 97 más barato: {gas_97[0][1]}. Más caro: {gas_97[-1][-1]}. Hay una diferencia de: {(gas_97[-1][-1])-(gas_97[0][1])}")
    archivo.write(f"\tGasolina 97 más barato: {gas_97[0][1]}. Más caro: {gas_97[-1][-1]}. Hay una diferencia de: {(gas_97[-1][-1])-(gas_97[0][1])}"+"\n")
    print(f"\tPetróleo diesel más barato: {diesel[0][1]}. Más caro: {diesel[-1][-1]}. Hay una diferencia de: {(diesel[-1][-1])-(diesel[0][1])}")
    archivo.write(f"\tPetróleo diesel más barato: {diesel[0][1]}. Más caro: {diesel[-1][-1]}. Hay una diferencia de: {(diesel[-1][-1])-(diesel[0][1])}"+"\n")
    print(f"\tGlp más barato: {glp[0][1]}. Más caro: {glp[-1][-1]}. Hay una diferencia de: {(glp[-1][-1])-(glp[0][1])}")
    archivo.write(f"\tGlp más barato: {glp[0][1]}. Más caro: {glp[-1][-1]}. Hay una diferencia de: {(glp[-1][-1])-(glp[0][1])}"+"\n")

    try:
        print(f"\tGnc más barato: {gnc [0][1]}. Más caro: es {gnc[-1][-1]}. Hay una diferencia de: {(gnc[-1][-1])-(gnc[0][1])}")
        archivo.write(f"\tGnc más barato: {gnc [0][1]}. Más caro: es {gnc[-1][-1]}. Hay una diferencia de: {(gnc[-1][-1])-(gnc[0][1])}"+"\n")
    except:
        print("\tNo se registra combustible gnc")
        archivo.write("\tNo se registra combustible gnc"+"\n")

except:
    print("nombre comuna no valida")
    archivo.write("nombre comuna no valida"+"\n")

archivo.close()
                      
```
