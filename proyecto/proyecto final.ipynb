```python
import requests

token = "hISHc9TmCE"
url = "http://api.cne.cl/v3/combustibles/vehicular/estaciones"
comuna = input('Ingrese la comuna: ').title()

res = requests.get(f"{url}?token={token}")
res_json = res.json()

archivo = open("archivo.txt", "w")

lista_precio = []
barato = []
caro = []

for d in res_json['data']:
    if d['id_region']=='13' and d['nombre_comuna']==comuna:
        print('Comuna:', d['nombre_comuna'])
        archivo.write("\n")
        archivo.write(f"Comuna: {d['nombre_comuna']}"+ "\n")
        print('\tFecha actualización:', d['fecha_hora_actualizacion'])
        archivo.write(f"\tFecha y hora de actualización: {d['fecha_hora_actualizacion']}"+ "\n")
        print('\tHorario de atención:', d['horario_atencion'])
        archivo.write(f"\tHorario de atención: {d['horario_atencion']}"+ "\n")
        print('\tNombre distribuidor:', d['distribuidor']['nombre'])
        archivo.write(f"\tNombre distribuidor: {d['distribuidor']['nombre']}"+ "\n")
        print('\tRazón Social:', d['razon_social'])
        archivo.write(f"\tRazón Social: {d['razon_social']}"+ "\n")
        print('\tDirección:', d['direccion_calle'], d['direccion_numero'])
        archivo.write(f"\tDirección: {d['direccion_calle']} {d['direccion_numero']}"+ "\n")
        for nombre, valor in (d['precios']).items():
            print(nombre, ':', valor)
            archivo.write(f"{nombre}: {valor}"+ "\n")
            lista_precio.append(valor)
            lista_precio.sort()
            if valor==lista_precio[0]:
                barato.append((valor, nombre))
                barato.sort()
            elif valor ==lista_precio[-1]:
                caro.append((valor,nombre))
                caro.sort()
        for met, res in (d['metodos_de_pago']).items():
            if res == False:
                res = 'No'
            else:
                res = 'Sí'
            print('\tMétodos de pago:', met, res)
            archivo.write(f"\t{met}: {res}"+ "\n")
        for s, servicio in (d['servicios']).items():
            if servicio == False:
                servicio = 'No'
            else:
                servicio = 'Sí'
            print('\tServicios:', s, servicio)
            archivo.write(f"\t{s}: {servicio}"+ "\n")
           
        print()



# conclusiones
archivo.write("\n")
print("=====================================================================================================")
archivo.write("====================================================================================================="+ "\n")
print('Basado en la información consultada, se puede concluir que: \n')
archivo.write("Basado en la información consultada, se puede concluir que:" +"\n")
print(f"\tEl combustible más barto de la comuna de {comuna} es {(barato[0][1])} y vale: {lista_precio[0]} pesos por litro.")
archivo.write(f"El combustible más barto de la comuna de {comuna} es {(barato[0][1])} y vale: {lista_precio[0]} pesos por litro." +"\n")
print(f"\tEl combustible más caro de la comuna de {comuna} es {caro[-1][-1]} y vale: {lista_precio[-1]} pesos por litro.")
archivo.write(f"El combustible más caro de la comuna de {comuna} es {caro[-1][-1]} y vale: {lista_precio[-1]} pesos por litro." +"\n")
print(f"\tLa diferencia entre el combustible más barto y más caro es: {lista_precio[-1] - lista_precio[0]} pesos.")
archivo.write(f"La diferencia entre el combustible más barto y más caro es: {lista_precio[-1] - lista_precio[0]} pesos." +"\n")
print()
print(f"\tEn 10 litros del combustible más barato ud gastará: {lista_precio[0]*10} pesos")
archivo.write(f"En 10 litros del combustible más barato ud gastará: {lista_precio[0]*10} pesos" +"\n")
print(f"\tEn 10 litros del combustible más caro ud gastará: {lista_precio[-1] * 10} pesos")
archivo.write(f"En 10 litros del combustible más caro ud gastará: {lista_precio[-1] * 10} pesos" +"\n")
print(f"\tLa diferencia es de : {(lista_precio[-1] * 10) - (lista_precio[0]*10)} pesos")
archivo.write(f"La diferencia es de : {(lista_precio[-1] * 10) - (lista_precio[0]*10)} pesos" +"\n")

archivo.close()

```
